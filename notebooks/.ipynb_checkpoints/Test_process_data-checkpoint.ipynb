```python
import iris
import numpy as np
import matplotlib.pyplot as plt
import iris.quickplot as qplt
import glob
import os, sys
%matplotlib
```

```python
data_dir = '/scratch/hadpx/cold_surge_monitoring/mogreps/raw_data/2020/11/01/12'
```

```python
files_members = glob.glob(os.path.join(data_dir, '*'))
files_members.sort()
members = [ff.split('/')[-1] for ff in files_members][:3]
members
```

```python
fc_times_all=np.arange(0, 192, 24)
```

```python
fc_times = fc_times_all#[4:5]
```

```python
np.where(fc_times_all==24)[0][0]
```

```python
precip_24hr_accum = []
for member in members:
    data_mem_dir = os.path.join(data_dir, member)
    realiz_coord = iris.coords.AuxCoord([int(member)], standard_name='realization', var_name='realization')
    for fc_time in fc_times:
        iloc = np.where(fc_time==fc_times_all)[0][0]    
        if iloc == 0:
            #print(iloc)
            filename_current = os.path.join(data_mem_dir, 'englaa_pd%s.pp' %str('%03d' % fc_times_all[iloc]))
            cube_curr = iris.load_cube(filename_current, 'precipitation_amount')
            #print(cube_curr.coord('time'))
            # 3 hourly sum for the analysis value
            # Not sure if this is accurate
            # check if the cubes are 3d
            if len(cube_curr.shape)==3:
                temp = cube_curr[-1].copy()
            else:
                print('HERE')
                temp = cube_curr.copy()
            # removing existing realization coord and adding a 
            # freshly cooked one
            if temp.coords('realization'):
                temp.remove_coord('realization')
            temp.add_aux_coord(realiz_coord)

            temp.data = (cube_curr[-1].data - cube_curr[0].data)*8.
            precip_24hr_accum.append(temp)
        else:
            #print(iloc)
            filename_previous = os.path.join(data_mem_dir, 'englaa_pd%s.pp' %str('%03d' % fc_times_all[iloc-1]))
            filename_current = os.path.join(data_mem_dir, 'englaa_pd%s.pp' %str('%03d' % fc_times_all[iloc]))
            #print(filename_current)
            #print(filename_previous)
            cube_prev = iris.load_cube(filename_previous, 'precipitation_amount')
            cube_curr = iris.load_cube(filename_current, 'precipitation_amount')
            
            # check if the cubes are 3d
            if len(cube_curr.shape)==3:
                temp = cube_curr[-1].copy()
            else:
                temp = cube_curr.copy()
            # removing existing realization coord and adding a 
            # freshly cooked one
            if temp.coords('realization'):
                temp.remove_coord('realization')
            temp.add_aux_coord(realiz_coord)
            
            # check if the cubes are 3d
            if len(cube_prev.shape)==3:
                temp.data -= cube_prev[-1].data
            else:
                print('Here')
                temp.data -= cube_prev.data
            
            precip_24hr_accum.append(temp)
            #print(precip_24hr_accum.data.min(), precip_24hr_accum.data.max())
            #qplt.plot(cube_prev[:,480,500])
```

```python
precip_24hr_accum
```

```python
cube = precip_24hr_accum[0]
realiz_coord = iris.coords.AuxCoord([3], standard_name='realization', var_name='realization')
if cube.coords('realization'):
    cube.remove_coord('realization')
cube.add_aux_coord(realiz_coord)
```

```python
cube.coord('realization')
```

```python
cubes = iris.cube.CubeList(precip_24hr_accum).merge_cube()
```

```python
if cubes.coord('time').bounds is not None:
    cubes.coord('time').bounds = None
    cubes.coord('time').guess_bounds()
    print(cubes.coord('time').bounds)
iris.util.promote_aux_coord_to_dim_coord(cubes, 'time')
```
